In this notebook I want to explore my dataset of declinable variants of Russian nouns that are indeclinable in standard language. Examples like this are studied: "_в длинном пальте_". The data were collected from General Internet Corpus of Russian(GICR)[http://www.webcorpora.ru/]. The list of studied nouns is based on the list used for study of gender variation of indeclinable nouns (Chuprinko,Magomedova,Slioussar (to appear)).
The query consisted of adjective and indeclinable noun with the last symbol substituted with the wildcard. There are two foci of interest: chosen declension type and assigned gender.
Not all searched nouns appeared with declinable variants in the corpus. In this notebook data of declinable and indeclinable variants are not compared, please take any references to indeclinable nouns for granted.

```python
import pandas as pd
import sqlite3
```

# 0. Download data and little prettifying

```python
#download data
df = pd.read_csv('declinable_variants.csv', sep=',', encoding='utf-8')
df.sample(4)
```

```python
# rename some columns
df = df.rename(columns={'result': "adjective",'Unnamed: 4':"noun_form", "source":"corpus", "id":"id_example"})
```

```python
# these are columns to delete, they are by-products of corpus search, they do not contain any information interesting to us
l_to_drop = ['int_id', 'center1', 'center2']

df = df.drop(l_to_drop, axis = 1)
df.head()
```

# 1. Database

```python
# I am planning to create three tables database via implementing pandas.dataframe.to_sql three times to three dataframes

#df = df.drop("int_id",axis = 1) #drop ids assigned inter queries

### create authors table
auth_l = ["author", "loc","birth"]
auth_df = df[auth_l]
auth_df = auth_df.drop_duplicates().reset_index(drop = True)
auth_df["id_author"] = auth_df.index 
auth_df.head()

# add author_id column to the phrase table
df = df.merge(auth_df[["author","id_author"]],  how = "left", on="author")
# образуется лишний id

### create source table
source_l = ["url", "year","genrei", "corpus", "corp_index", "id_example", "compose_doc_id","sub_query_index","group_id"]
source_df = df[source_l]
source_df = source_df.drop_duplicates().reset_index(drop = True)
source_df["id_source"] = source_df.index 

# add source_id column to the phrase table
df = df.merge(source_df[["url","compose_doc_id","id_source"]],  how = "left", on=["url","compose_doc_id"])

### drop columns in the phrase table that were distributed among other tables
df = df.drop(source_l, axis = 1)
df = df.drop(auth_l, axis = 1)

# rename table and watch result
phrase_df = df.copy()
phrase_df["id_phrase"] = phrase_df.index
phrase_df.sample(4)
```

```python
con = sqlite3.connect("declinable_variants.db");
cur = con.cursor();
phrase_df.to_sql(name='phrases', con=con, if_exists='replace');
source_df.to_sql(name='sources', con=con, if_exists='replace');
auth_df.to_sql(name='authors', con=con, if_exists='replace');
```

```python
# check if it worked well
for row in cur.execute('SELECT * FROM phrases'):
    print(row)
```

# 2. Dataframe creation and data manipulation

```python
## I want to work with phrases and authors tables

query = """
SELECT *
FROM phrases 
    JOIN authors ON phrases.id_author = authors.id_author
WHERE birth > ? AND loc IS NOT NULL
"""
# birth in WHERE clause is used to drop unrealistic old people
```

```python
birth = 1900
data = pd.read_sql_query(query, params = [birth], con=con)
data = data.drop(['id_author','id_source','id_phrase','index'], axis =1 )
data.sample(4)
```

We want to test how NLP instruments will work with such misuses like 'в пальте'. Will it recognize intended nouns?

First, let's try mystem lemmatizer

```python
import pymystem3
```

```python
m = pymystem3.Mystem()
mystem_result = []

for word in data.noun_form:
    x = m.lemmatize(word)
    x.remove('\n')
    x = "".join(x) # for words with "-" like "фуа-гра"
    mystem_result.append(x)
data['mystem_result'] = mystem_result
print(data['mystem_result'][25:45]) # see some results
```

Obviously, mystem did not work well.

Second, let's try pymorphy2 parser

```python
from pymorphy2 import MorphAnalyzer
morph = MorphAnalyzer()
```

```python
pymorphy_result = []

for word in data.noun_form:

    x = morph.parse(word)[0][2]
    pymorphy_result.append(x)
data['pymorphy_result'] = pymorphy_result
```

```python
data.sample(5)
```

As we can see in many cases both instruments have not worked well.
Pymorphy's perofrmance seems to be slightly better. In its view "в чёрном пальте" contains the noun "пальт" what is somewhat closer than verb "палить" introduced by mystem.

```python
import re 
# delete everything before the noun and delete "склон." + extensions
begin = re.compile(r'^.+ \+ ')
end = re.compile(r' \(склон\.\)\.(xlsx|xls)')

word_l = []
for file in list_files:
    if "0" in file:
        file = begin.sub(" ", file)
        file = end.sub(" ", file)
        word_l.append(file.strip())
```

I am planning to use Levenshtein distance _Levenshtein_ to find similarity between "misused" forms and forms in list of implied nouns. Then we will automatically assign them in the dataframe.

```python
# pip install python-Levenshtein # https://pypi.org/project/python-Levenshtein/ #https://maxbachmann.github.io/Levenshtein/levenshtein.html#distance
from Levenshtein import distance

actual_words = []
for misuse in data.noun_form:
    d = {}
    for true_word in word_l:
        dist = distance(misuse,true_word)
        d[true_word] = dist
        least_dist_w = sorted(d.items(), key=lambda x: x[1], reverse = False)[0][0]
    actual_words.append(least_dist_w)

data["actual_words"] = actual_words
```

```python
data.sample(5)
```

Let's automatically extract gender/plurality features from left context adjectives. The parsers do not extract declension features (nor theoretical linguists agree on classification). We will simply assume that gender and declension are strongly correlated.

```python
gender = []
for adj in data.adjective:
    feature = str(morph.parse(adj)[0][1])
    if "plur" in feature:
        gender.append("pl")
    elif "femn" in feature:
        gender.append("fem")
    elif "masc" in feature:
        gender.append("masc")
    elif "neut" in feature:
        gender.append("neut")
    else: # in case if some word without gender or number category appeared in the left context
        gender.append("neut")
        
data["gender"] = gender
# masculine and neuter can be united, since most of the examples are actually ambigious between both of them due to Russian case syncretism
data = data.replace({"masc":"masc/neut","neut":"masc/neut"})
```

Let us extract last symbols of the intended nouns into a new column. It is the feature that will allow to test us if indeclinable nouns transform into declinable variants by analogy with the original final vowel or some other patterns including vowels substitution take place.

```python
vowels = []
for noun in data.actual_words:
    last_symbol = noun[-1]
    vowels.append(last_symbol)
data['vowels'] = vowels
data = data.replace({"я":"а"}) #since "я" group is so small we can consider it as "а"
```

Let's add some synctatic analysis. **Objective**: extract dependency of the noun form.

```python
import spacy

#! python -m spacy download ru_core_news_sm

nlp = spacy.load("ru_core_news_sm")
```

```python
#let's replace "None" values in right and left contexts with empty strings
data = data.replace({None: ""})

data['sentence'] = data['left'] + " " + data['adjective']  + " " + data['noun_form'] + " " + data['right']
```

```python
dependency_l = []

for sent, form in zip(data.sentence, data.noun_form):
  analyze = nlp(str(sent))
  if form in str(analyze):
    #print(form)
    splitted = str(analyze).split(" ")
    index = splitted.index(form)
    dependency_l.append(analyze[index].dep_)

data["dependency"] = dependency_l
```

```python
frequencies_df = data.groupby(['vowels',"gender"])['gender'].agg("count").to_frame(name = "counts")
frequencies_df = frequencies_df.reset_index()
frequencies_df
```

```python
# pie chart for genders in the dataset in general
frequencies_df2 = frequencies_df.groupby(['gender']).agg('sum').reset_index()
frequencies_df2

plt.figure(figsize=(6, 6))
frequencies_df2['gender'].value_counts().plot(kind='pie',legend=True, title='Air Termination System', ylabel='',labeldistance=None);
plt.title('Gender');
```

**Comments:**
In this pie chart we can see relative gender frequencies in total in our data. Overall we observe equal shares of all genders given all the nouns. 

```python
# I want to make log transformation because some nouns are more frequent in our data while others are rarer
frequencies_df['counts_log'] = np.log(frequencies_df['counts'])
log_frequencies_df = frequencies_df.drop('counts', axis = 1).groupby(["vowels",'gender']).agg('sum').reset_index()
```

```python
# heatmap
vowel_gender_matrix = log_frequencies_df["counts_log"].to_numpy().reshape(5,3) # create matrix

sns.heatmap(
    vowel_gender_matrix,
    annot=True,
    xticklabels=['fem','masc/neut','pl'],
    yticklabels=['а','е','и','о','у'],
    cbar_kws={'label': 'log Gender frequencies'}
)
plt.xlabel('Genders/plurality')
plt.ylabel('Ending vowels of the meant nouns')
plt.show()
```

**Comments:**
Overall the most frequent genders among each group were predictable. _-а_ indeclinable nouns like "фуа-гра" are similar to feminine gendered declinables and *-и* nouns are similar to plural forms.
The pervasiveness of feminine gender and corresponding declension seems untrivial. Usually feminine is considered as marked feature so it is oftenly rare, nevertheless, in our data it is quite frequent for all types of nouns. Working hypothesis: native speakers feel how rare the feminine is and how it is incompatible with indeclinables, thus, they use it ironically to mock the nouns more effectively.

***CONCLUSIONS***
In this notebook we investigated behaviour of declinable misused form of the Russian indeclinable nouns. So far not so much was found, but gender distribution among groups of nouns with different ending vowels seemed peculiar. On the one side, it is intuitive: indeclinable nouns are converting to similar declinable nouns. On the other hand, somewhat discourse oriented pattern is observable, share of feminine gender usage among all nouns is surprisingly large. Current hypothesis is that native speakers intentionally use feminine gender to make fun of the foreign indeclinable 
word. 

The exploratory analysis did not hint gender preferences in combination with specific syntactic position.
